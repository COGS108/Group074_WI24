```python
life_expectancy = pd.read_csv("./datasets/life_expectancy.csv")

life_expectancy.columns = life_expectancy.iloc[3]        # setting the column names correctly
life_expectancy = life_expectancy.iloc[4:]               # remove extra rows
life_expectancy = life_expectancy.reset_index(drop=True) # Reset the indices
life_expectancy = life_expectancy.dropna(axis=1, how='all')
del life_expectancy['Indicator Code']
del life_expectancy['Indicator Name']

life_expectancy.columns = life_expectancy.columns.astype(str)
life_expectancy = life_expectancy.filter(regex='^(?!19\d\d).*$').rename(columns=lambda x: str(x.split('.')[0]))
```
