### Dataset Cleaning -- BMI

Following the cleaning steps, we removed the irrelevant row names and formating the data.


```python
bmi = pd.read_csv("./datasets/bmi_country.csv")
bmi.shape
```

```python
# CLEANING UP TABLE
bmi = bmi[bmi['Year'] >= 2000] # updating it so that it is only in 21st century
if 'ISO' in bmi.columns: # deletes NaN column that wasn't used
    del bmi['ISO']

# the regions names that need to be removed
to_remove = ['Africa', 'Asia', 'Europe', 'European Union', 
             'income', 'situations','South America', 'North America',
             'Asia', 'High-income countries', 'Low-income countries',
             'USSR', 'World', 'Oceania', 'demograph', 'poor', 'total',
             'Not classified', 'Least developed countries: UN classification',
             "St. Martin (French part)", "Korea, Dem. People's Rep.", 
             "British Virgin Islands", "Gibraltar", "IBRD only", "IDA blend",
             "IDA", "OECD members", "Other small states"]
bmi = bmi[~bmi['Country/Region/World'].str.contains('|'.join(to_remove), case=False)]
bmi = bmi.reset_index(drop=True)
```

```python
# cleans up uncertainty interval, we can add back if needed for EDA but not sure
if 'uncertainty interval' in bmi.columns:
    del bmi['uncertainty interval']
```

```python
# Setting up the row names, col names, and formatting the dataset
if 'Sex' in bmi.columns:
    bmi = bmi.drop("Sex", axis=1)
bmi.loc[bmi['Country/Region/World'] == 'United States of America', 'Country/Region/World'] = "United States"
grouped_bmi = bmi.groupby(['Country/Region/World', 'Year']).mean()
mean_only = grouped_bmi.loc[:, grouped_bmi.columns == 'Mean BMI']
mean_only = mean_only.reset_index()
bmi_pivot = mean_only.pivot_table(index=['Country/Region/World'], 
                                columns='Year', 
                                values='Mean BMI')
bmi_pivot.insert(0, 'Series', 'BMI')
bmi_pivot.columns.name = None
bmi_pivot = bmi_pivot.reset_index()
bmi_pivot = bmi_pivot.rename(columns = {'Country/Region/World': 'Country'})
```

### Data Cleaning -- Fat

Following the cleaning steps, we removed the irrelevant row names and formatting the data.

```python
fat_supply = pd.read_csv("./datasets/daily-per-capita-fat-supply.csv")
fat_supply.shape
```

```python
# The list of region names that need to be removed
group_delete_name = ['Africa', 'Asia', 'Europe', 'European Union', 
             'income', 'situations','South America', 'North America',
             'Asia', 'High-income countries', 'Low-income countries',
             'USSR', 'World', 'Oceania', 'demograph', 'poor', 'total',
             'Not classified', 'Least developed countries: UN classification',
             "St. Martin (French part)", "Korea, Dem. People's Rep.", 
             "British Virgin Islands", "Gibraltar", "IBRD only", "IDA blend",
             "IDA", "OECD members", "Other small states"]

# created functions that help formatting the country names
def clean_FAO(string):
    if '(FAO)' in string or '(27)' in string:
        return False
    else:
        return True

def clean_former(string):
    if ('(former)' in string):
        return string.replace(' (former)', '')
    elif ('(country)' in string):
        return string.replace(' (country)', '')
    else:
        return string
def clean_others(string):
    if string in group_delete_name:
        return False
    else:
        return True
def clean_code(string):
    if string == np.nan:
        return string.replace(np.nan, 'SDN')
    else:
        return string

fat_supply = fat_supply.dropna(axis=0, how='all')
fat_supply = fat_supply[fat_supply.get('Year') >= 2000]
fat_supply.columns = ['Country', 'Code', 'Year', 'Fat_Per_capita']
fat_supply = fat_supply[fat_supply.get('Country').apply(clean_FAO)]
fat_supply['Country'] = fat_supply['Country'].apply(clean_former)
fat_supply = fat_supply[fat_supply.get('Country').apply(clean_others)]
fat_supply = fat_supply.fillna('SDN')

# add the series column 
pivot_df = fat_supply.pivot(index='Country', columns=['Year'], values='Fat_Per_capita')
pivot_df.insert(0, 'series','Fat (g/person/day)')
fat_supply2 = pivot_df
```

Based on the provided OLS regression results, we can interpret them step by step:

##### Overall Model Significance:

The low p-value indicates that the overall model is statistically significant.

The R-squared value is 0.406, suggesting that approximately 40.6% of the variance in the dependent variable (BMI) is explained by the independent variables in the model.
The Adjusted R-squared is 0.394, which adjusts for the number of predictors in the model.

The intercept (const) has a coefficient of 20.7978 with a low p-value (< 0.0001), indicating that it is statistically significant.
Carbs (g/day) has a coefficient of 0.0035 with a p-value of 0.245, which is not statistically significant at the conventional significance level of 0.05.

Protein (g/day) has a coefficient of -0.0024 with a p-value of 0.868, indicating that it is not statistically significant.

Fat (g/day) has a coefficient of 0.0381 with a low p-value (< 0.0001), suggesting that it is statistically significant.

The intercept represents the expected BMI when all independent variables are zero. In this case, it's 20.7978.

For each additional gram of fat consumed per day, the BMI is expected to increase by 0.0381 units, holding other variables constant.

Carbs and protein intake do not have statistically significant effects on BMI in this model.

##### Multicollinearity:

The condition number is very large (3.56e+03), which may indicate potential multicollinearity issues among the independent variables.
Residuals:

The Omnibus and Jarque-Bera tests provide information about the normality of residuals. In this case, p-values for both tests are above 0.05, suggesting that the residuals may be normally distributed.

##### Overall Assessment:

The model overall is statistically significant, but only the consumption of fat per day appears to have a significant effect on BMI in this model. However, caution should be exercised due to potential multicollinearity issues and the non-significance of other predictors.

## Proportional Data
To find the magnitude of the impact of what macro affects life expectancy/bmi the most, we should at how countries diets eat proportionally more of 1 macro versus another. From there, we can find the top countries that have the highest relative proportion to the world of each macro, to see if eating a proportionally more amount of one macro can positively/negatively affect someone's life. The below code calculates the proportions of macros that each country consumes per capita, separated by fats, carbs, protein.

```python
nutrients_data = averages[averages['Series'].isin(['Fat (g/day)', 'Protein (g/day)', 'Carbs (g/day)'])]

# Pivot the table to have one row per country with separate columns for fat, protein, and carbs
pivot_data = nutrients_data.pivot(index='Country', columns='Series', values='Average over 2000 - 2016')

# Calculate the ratios
pivot_data['Total Intake'] = pivot_data.sum(axis=1)
pivot_data['Fat Ratio'] = pivot_data['Fat (g/day)'] / pivot_data['Total Intake']
pivot_data['Carbs Ratio'] = pivot_data['Carbs (g/day)'] / pivot_data['Total Intake']
pivot_data['Protein Ratio'] = pivot_data['Protein (g/day)'] / pivot_data['Total Intake']

# Create the new dataset with the requested list format
new_dataset = pivot_data[['Fat Ratio', 'Carbs Ratio', 'Protein Ratio']].apply(lambda x: [x['Fat Ratio'], x['Carbs Ratio'], x['Protein Ratio']], axis=1)
new_dataset = new_dataset.reset_index()
new_dataset.columns = ['Country', 'Ratios (Fats, Carbs, Protein)']

# Display the first few rows of the new dataset to verify
new_dataset.head()
```

Next we want to compare the proportions to the average global proportion.

```python
prop_global = global_avg.copy()
columns_of_interest = list(range(2000, 2017))
prop_global['Average over 2000 - 2016'] = prop_global.loc[:, columns_of_interest].mean(axis=1)
prop_global = prop_global.drop(columns=columns_of_interest)
prop_global


total = prop_global["Average over 2000 - 2016"][1] + prop_global["Average over 2000 - 2016"][2] + prop_global["Average over 2000 - 2016"][4]
carbs1 = prop_global["Average over 2000 - 2016"][1]/total
fat1 = prop_global["Average over 2000 - 2016"][2]/total
protein1 = prop_global["Average over 2000 - 2016"][4]/total

global_ratios = [fat1, carbs1, protein1]
```

```python
prop_difference = new_dataset.copy()
prop_difference['Difference of Ratios (Fats, Carbs, Protein), vs Global'] = prop_difference['Ratios (Fats, Carbs, Protein)'].apply(lambda x: [i - r for i,r in zip (x, global_ratios)])
if 'Ratios (Fats, Carbs, Protein)' in prop_difference.columns:
    prop_difference.drop(columns='Ratios (Fats, Carbs, Protein)', inplace=True)
prop_difference
# Separate the items into three separate columns
prop_difference[['Proportional Difference (PD) in Fats', 'PD in Carbs', 'PD in Protein']] = prop_difference['Difference of Ratios (Fats, Carbs, Protein), vs Global'].apply(pd.Series)

# Drop the original column with lists
prop_difference.drop(columns=['Difference of Ratios (Fats, Carbs, Protein), vs Global'], inplace=True)
prop_difference.dropna(inplace = True)
```
