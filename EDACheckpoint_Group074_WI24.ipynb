# Research Question

How do the various macronutrients (carbs as sugar, protein, fat) primarily consumed by individuals in various countries impact BMI rates in the twenty first century?

## Background and Prior Work

Based on different geographical locations, people have access to different resources, climates, and agricultural commodities. Therefore, this proposed study seeks to examine the relationship between regional diets and health metrics such as weight and longevity.

### Data Cleaning -- daily-per-capita-fat-supply.csv

```python
fat_supply = pd.read_csv("./datasets/daily-per-capita-fat-supply.csv")
fat_supply.shape
```

```python
group_delete_name = ['Africa', 'Asia', 'Europe', 'European Union', 
             'income', 'situations','South America', 'North America',
             'Asia', 'High-income countries', 'Low-income countries',
             'USSR', 'World', 'Oceania', 'demograph', 'poor', 'total',
             'Not classified', 'Least developed countries: UN classification',
             "St. Martin (French part)", "Korea, Dem. People's Rep.", 
             "British Virgin Islands", "Gibraltar", "IBRD only", "IDA blend",
             "IDA", "OECD members", "Other small states"]
def clean_FAO(string):
    if '(FAO)' in string or '(27)' in string:
        return False
    else:
        return True

def clean_former(string):
    if ('(former)' in string):
        return string.replace(' (former)', '')
    elif ('(country)' in string):
        return string.replace(' (country)', '')
    else:
        return string
def clean_others(string):
    if string in group_delete_name:
        return False
    else:
        return True
def clean_code(string):
    if string == np.nan:
        return string.replace(np.nan, 'SDN')
    else:
        return string

fat_supply = fat_supply.dropna(axis=0, how='all')
fat_supply = fat_supply[fat_supply.get('Year') >= 2000]
fat_supply.columns = ['Country', 'Code', 'Year', 'Fat_Per_capita']
fat_supply = fat_supply[fat_supply.get('Country').apply(clean_FAO)]
fat_supply['Country'] = fat_supply['Country'].apply(clean_former)
fat_supply = fat_supply[fat_supply.get('Country').apply(clean_others)]
fat_supply = fat_supply.fillna('SDN')
fat_supply.head()

pivot_df = fat_supply.pivot(index='Country', columns=['Year'], values='Fat_Per_capita')
pivot_df.insert(0, 'series','Fat (g/person/day)')
fat_supply2 = pivot_df
```
