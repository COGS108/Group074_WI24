```python
# CLEANING UP TABLE
bmi = bmi[bmi['Year'] >= 2000] # updating it so that it is only in 21st century
if 'ISO' in bmi.columns: # deletes NaN column that wasn't used
    del bmi['ISO']
bmi.head()
```

```python
bmi.shape
```

Next up, standardddizing the table, with sex and relevant BMI information

```python
def standardize_gender(str_in): 
    try:
        if pd.notna(str_in):
            str_in = str_in.lower().strip()

            
            if 'women' in str_in:
                return str_in.replace('women', 'F')
            elif 'men' in str_in:
                return str_in.replace('men', 'M')
            else:
                return np.nan
        else:
            return np.nan
    except Exception as e:
        print(f"An error occurred: {e}")
        return np.nan
```

```python
bmi['Sex'] = bmi['Sex'].apply(standardize_gender)
```

```python
# cleans up uncertainty interval, we can add back if needed for EDA but not sure
for col in bmi.columns:
    if 'uncertainty interval' in col:
        del bmi[col]
bmi
```

```python
bmi.to_csv('./cleaned_datasets/bim_country_clean.csv', index=False)
```

### Data Cleaning -- daily-per-capita-fat-supply.csv

```python
fat_supply = pd.read_csv("./datasets/daily-per-capita-fat-supply.csv")
fat_supply.shape
```

```python
group_delete_name = ['Africa', 'Aisa', 'Europe', 'High-income countries', 'Low-income countries', 'Lower-middle-income countries', 'North America', 'South America', 'Upper-middle-income countries', 'World', 'Oceania']
def clean_FAO(string):
    if '(FAO)' in string or '(27)' in string:
        return False
    else:
        return True

def clean_former(string):
    if ('(former)' in string):
        return string.replace(' (former)', '')
    elif ('(country)' in string):
        return string.replace(' (country)', '')
    else:
        return string
def clean_others(string):
    if string in group_delete_name:
        return False
    else:
        return True

fat_supply = fat_supply.dropna(axis=0, how='all')
fat_supply = fat_supply[fat_supply.get('Year') >= 2000]
fat_supply.columns = ['Country', 'Code', 'Year', 'Fat_Per_capita']
fat_supply = fat_supply[fat_supply.get('Country').apply(clean_FAO)]
fat_supply['Country'] = fat_supply['Country'].apply(clean_former)
fat_supply = fat_supply[fat_supply.get('Country').apply(clean_others)]
fat_supply
```

```python
fat_supply.to_csv('./cleaned_datasets/fat_per_capita_clean.csv', index=False)
```
